## Sravni

```python
sravni = pd.read_csv(data_path + 'sravni_reviews.csv').drop(columns='Unnamed: 0')
```

```python
ur_products = [
    'mobilnoyeprilozheniye', 'rko', 'ekvajring', 
    'businessRko', 'acquiring'
]
```

```python
sravni = sravni[~sravni['tag'].isin(ur_products)]
```

```python
sravni['tag'].value_counts()
```

```python
sravni = sravni.reset_index()
```

```python
sravni.rename(columns={'date' : 'dateCreate', 'tag' : 'product'}, inplace=True)
```

```python
sravni.rename(columns={'specificProductName' : 'sub_product'}, inplace=True)
```

```python
sravni
```

```python
sravni['sub_product'].value_counts()
```

```python
banki['sub_product'] = 'Отсутствует'
```

```python
import numpy as np

sravni_new = sravni[["id", "title", "text", 'dateCreate', 'product', 'sub_product']].copy()
sravni_new["source"] = "sravni"

banki_new = banki[["id", "title", "text", 'dateCreate', 'product', 'sub_product']].copy()
banki_new["source"] = "banki"

merged = pd.concat([sravni_new, banki_new], ignore_index=True)
```

```python
merged = merged.reset_index()
```

```python
merged['sub_product'] = merged['sub_product'].fillna('Отсутствует')
```

```python
merged
```

```python
merged['product'].unique()
```

```python
mapping = {
    "debitcards": "debitCards",
    "debitCards": "debitCards",

    "creditcards": "creditCards",
    "creditCards": "creditCards",

    "hypothec": "mortgage",
    "mortgage": "mortgage",

    "deposits": "savings",
    "savings": "savings",

    "transfers": "moneyOrder",
    "moneyOrder": "moneyOrder",
}
```

```python
merged["product"] = merged["product"].replace(mapping)
```

```python
merged.to_csv('../data/prepared_data/prepared_reviews_v4(classes).csv', index = False)
```
